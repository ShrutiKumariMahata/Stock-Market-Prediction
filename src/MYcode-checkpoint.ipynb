```python
import numpy as np     #numpy (np): For numerical operations and arrays
import pandas as pd   #pandas (pd): For handling tabular data (DataFrames)
import matplotlib.pyplot as plt    #matplotlib.pyplot (plt): For plotting graphs
import yfinance as yf   #yfinance (yf): To download stock market data from Yahoo Finance
```

```python
start='2012-01-01'
end='2022-12-21'
stock='GooG'
data=yf.download(stock,start,end)
```

```python
data.reset_index(inplace=True)  #Resets the index of the DataFrame data to default integer index (0, 1, 2, …).
```

```python
data.dropna(inplace=True)   # delete null and missing values perticular data will be deleted
```

```python
data_train = pd.DataFrame(data.Close[0:int(len(data)*0.80)])#First line: takes the first 80% of closing prices to train the LSTM model.
data_test = pd.DataFrame(data.Close[int(len(data)*0.80):len(data)])   #second line: takes the remaining 20% to test the model on unseen data.
```

```python
data_train.shape[0]  #Returns the number of rows (data points / days) in data_train.  shape → gives (rows, columns) [0] → selects rows
```

```python
data_test.shape[0]   #Returns the number of rows (data points) in the test dataset.
```

```python
from sklearn.preprocessing import MinMaxScaler #Imports the MinMaxScaler class from scikit-learn, which is used to normalize data.
scaler = MinMaxScaler(feature_range=(0,1))  #Creates a scaler object that rescales data between 0 and 1, which is important for neural networks like LSTM to work efficiently and avoid training issues.
```

```python
data_train_scale = scaler.fit_transform(data_train)   #fit → learns the minimum and maximum values from data_train

#transform → scales the data to the range 0–1
#Result: data_train_scale is the normalized training data, ready for LSTM training
```

```python
x = []   #Empty lists to store input sequences (X) and target values (Y).
y = []

for i in range(100, data_train_scale.shape[0]):   #Loop starts from index 100 to create sequences of 100 previous days.
    x.append(data_train_scale[i-100:i])    #Takes the last 100 data points as one input sequence for LSTM.
    y.append(data_train_scale[i,0])          #Stores the next day’s price as the output to predict.
```

```python
x, y = np.array(x), np.array(y)   #Converts the Python lists x and y into NumPy arrays.
```

```python
from keras.layers import Dense,Dropout,LSTM            #LSTM: Learns patterns from time-series data (stock prices).
#Dropout: Prevents overfitting by randomly dropping neurons during training.
#Dense: Fully connected layer used for final prediction.
from keras.models import Sequential   #Sequential: Builds the model layer-by-layer in order.
```

```python
model = Sequential()   #Initializes a Sequential model (layers are added one after another).
model.add(LSTM(units = 50, activation = 'relu' , return_sequences = True,
               input_shape = ((x.shape[1],1))))
model.add(Dropout(0.2))        #LSTM with 50 units
                                  #activation='relu' → ReLU activation function
                                 #return_sequences=True → returns the full sequence (needed for next LSTM layer)
                                         #input_shape=(x.shape[1],1) → input has 100 timesteps and 1 feature
                               #Dropout 20% of neurons to prevent overfitting

model.add(LSTM(units = 60, activation='relu', return_sequences = True))
model.add(Dropout(0.3))           #LSTM with 60 units  ,  Dropout 30%

model.add(LSTM(units = 80, activation='relu', return_sequences = True))
model.add(Dropout(0.4))     #LSTM with 80 units,  Dropout 40%

model.add(LSTM(units = 120, activation='relu', return_sequences = True))
model.add(Dropout(0.5))            # LSTM with 120 units ,  Dropout 50%

model.add(Dense(units=1))            #Dense layer with 1 neuron to predict the next day stock price
```

```python
model.compile(optimizer = 'adam',loss = 'mean_squared_error')    #optimizer='adam' → Uses the Adam optimizer to update model weights efficiently during training.
#loss='mean_squared_error' → Measures the difference between predicted and actual prices; the model tries to minimize this error.
```

```python
model.fit(x,y, epochs = 50,batch_size = 32, verbose =1)   #x, y → input sequences and target prices

#epochs=50 → model trains 50 times over the entire dataset
#batch_size=32 → updates weights after 32 samples
#verbose=1 → shows training progress
```

```python
model.summary()   #Displays a summary of the model architecture, including:

#Layers and their types (LSTM, Dense, Dropout)
#Output shapes of each layer
#Number of trainable parameters
```

```python
pas_100_days = data_train.tail(100)   #Selects the last 100 rows of the training data.
```

```python
data_test = pd.concat([pas_100_days, data_test], ignore_index=True)  #Combines the last 100 training days (pas_100_days) with the test dataset.
```

```python
data_test_scale = scaler.fit_transform(data_test)  #Normalizes the test dataset using MinMaxScaler (scales values between 0 and 1).
```

```python
x = []
y = []

for i in range(100, data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])
x, y = np.array(x), np.array(y)
```

```python
y_predict = model.predict(x)  #Uses the trained LSTM model to predict stock prices based on input sequences x.
```

```python
scale = 1/scaler.scale_  #Calculates the inverse of the scaling factor used by MinMaxScaler.
```

```python
y_predict = y_predict*scale  #Converts the predicted values back from the normalized (0–1) scale to the original stock price scale.
```

```python
y = y*scale  #Converts the actual target values from the normalized scale back to the original stock price scale.
```

```python
plt.figure(figsize=(10,8))  #Creates a plot with size 10 × 8 inches.
plt.plot(y_predict,'r',label = 'Predicted Price')  #Plots the predicted stock prices in red.
plt.plot(y,'g',label = 'Original Price') #Plots the actual stock prices in green.
plt.xlabel('Time')  #Labels the x-axis as "Time" (represents days or data points).
plt.ylabel('Price')   #Labels the y-axis as "Price" (stock price values).
plt.legend()  #Displays the legend for the plotted lines (e.g., Predicted vs Actual).
plt.show()   #Displays the plot on the screen.
```
